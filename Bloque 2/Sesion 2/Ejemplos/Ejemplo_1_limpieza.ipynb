# Ejemplo_1_limpieza

Los datos, en su forma natural (lo que llamamos “Raw Data”), suelen tener errores de registro que imposibilitan un análisis exacto. Al ser registrados por distintos sistemas y personas, es normal que terminemos con un fichero en el que un mismo valor esté expresado de distintas maneras (por ejemplo, una fecha puede estar registrada como 28 de Junio, o como 28/06 en una mismo archivo), pueden haber registros en blanco, números con coma en vez de punto, etc.

Al momento de hacer un análisis de esos datos, todos esos registros tienen que pre-procesarse. Es decir, se tiene que limpiar, unificar, consolidar y normalizar los datos para que se puedan utilizar y lograr extraer información de valor (Data Wrangling).

Data, in its natural form (what we call “Raw Data”), often has registry errors that make exact analysis impossible. As they are registered by different systems and people, it is normal that we end up with a file in which the same value is expressed in different ways (for example, a date can be registered as June 28, or as 06/28 in the same file ), there may be blank records, numbers with commas instead of docs, etc.

When doing an analysis of that data, all those records have to be pre-processed. That is, the data has to be cleaned, unified, consolidated and normalized so that it can be used and extract valuable information (Data Wrangling).
