## Conexión a la BBDD

```python
#Generamos la cadena de conexión
conection_string = r'chinook.db'

# Creamos un cursor para las consultas
cnx = sqlite3.connect(conection_string)
```

### Consultas de una tabla

```python
df_tracks = pd.read_sql_query("SELECT * FROM tracks", cnx)
```

```python
df_customers = pd.read_sql_query("SELECT * FROM customers", cnx)
```

```python
df_customers_name = pd.read_sql_query("SELECT CustomerId,FirstName, LastName FROM customers", cnx)
```

```python
df_customers_canada = pd.read_sql_query("SELECT * FROM customers WHERE Country == 'Canada'", cnx)
```

```python
df_customers_fran = pd.read_sql_query("SELECT * FROM customers WHERE FirstName Like 'Fran%'", cnx)
```

### Consultas entre varias tablas

```python
df_customers_join = pd.read_sql_query("SELECT FirstName, LastName, Company, BillingAddress, InvoiceDate FROM customers INNER JOIN invoices ON customers.CustomerId = invoices.CustomerId", cnx)
```
