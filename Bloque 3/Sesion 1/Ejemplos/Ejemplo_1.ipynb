## 3. Función apply

### 3.1 Definición estandar

```python
# defining function to check price 
def fun(num): 
  
    if num<200: 
        return "Low"
  
    elif num>= 200 and num<400: 
        return "Normal"
  
    else: 
        return "High"
```

```python
# passing function to apply and storing returned series in new 
stock["score"] = stock["close"].apply(fun) 
```

```python
stock
```

### 3.2 Definición mediante ***lambda***

```python
stock["open"] = stock["open"].apply(lambda num : num + 5) 
```

```python
stock
```

```python
stock['high'] = stock['high'].apply(lambda x: np.nan if x < 90 else x)
```

```python
stock
```

```python
stock['low'] = stock['low'].apply(lambda x: int(2*math.sqrt(x)))
```
