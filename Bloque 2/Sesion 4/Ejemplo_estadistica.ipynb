## 8. Desciación Estandar

```python
customers_plot = customers[(customers["State"] == 'KS')|(customers["State"] == 'OH')|(customers["State"] == 'NJ')|(customers["State"] == 'OK')]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(style="white", color_codes=True)
ax = sns.stripplot(x="Day Mins", data=customers_plot)
ax.axvline(x=customers["Day Mins"].mean() + customers["Day Mins"].std(), linewidth=4, color='coral')
ax.axvline(x=customers["Day Mins"].mean() - customers["Day Mins"].std(), linewidth=4, color='coral')
ax.axvline(x=customers["Day Mins"].mean(), linewidth=4, color='coral')
#ax.text(5, 250, 'D. Estandar', fontsize=30, color='black', bbox={'facecolor': 'lightblue', 'alpha': 1, 'pad': 10})
ax.set_title('Frecuencia - Minutos diarios', fontsize=18)
ax.text(110, 0.55, '-Sigma', fontsize=10, color='red', weight = 'bold')
ax.text(220, 0.55, '+Sigma', fontsize=10, color='red', weight = 'bold')
ax.text(170, 0.55, 'Media', fontsize=10, color='red', weight = 'bold')
ax.text(5, -0.4, 'D. Estandar', fontsize=30, color='black', bbox={'facecolor': 'lightblue', 'alpha': 1, 'pad': 10})
ax.set_xlabel('Minutos diarios', fontsize=16, weight = 'bold')
ax.set_ylabel('Frecuencia', fontsize=16, weight = 'bold')
fig.set_size_inches(11, 7)
sns.despine()
```

## 9. Anomalías

```python
KS_customers = customers[customers["State"] == 'KS']["Day Mins"]
```

```python
Q1 = KS_customers.quantile(0.25)
Q3 = KS_customers.quantile(0.75)
iqr = Q3 - Q1
lower_bound = Q1 -(1.5 * iqr) 
upper_bound = Q3 +(1.5 * iqr) 
print("Las anomalías estan por debajo de: " + str(lower_bound) + " y están por encima de: " + str(upper_bound))
```

```python
outliers_low = KS_customers[KS_customers < lower_bound]
```

```python
outliers_low
```

```python
outliers_hi = KS_customers[KS_customers > upper_bound]
```
