## 1 Crear DataFrame

### 1.1 DataFrame a partir de dictado

```python
pd.DataFrame({'numbers':[1,2,3], 'colors':['red', 'white', 'blue']})
```

### 1.2 DataFrame a partir de numpy

```python
x = np.random.standard_normal(3)
```

```python
y = np.arange(3)
```

```python
pd.DataFrame({'X':x, 'Y':y})
```

### 1.3 DataFrame a partir de series de Pandas 

```python
serie_1 = pd.Series(np.random.randn(3), index = ['a', 'b', 'c'])
```

```python
serie_2 = pd.Series([100, 230, 42], index = ['a', 'b', 'c'])
```

```python
pd.DataFrame([serie_1, serie_2])
```
